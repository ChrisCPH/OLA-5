```python
# Loading data from pdf
import pdfplumber

pdf_texts = []

pdf_files = ["./data/Rapport-Evidensgrundlaget_for_danske_raad_om_kost_og_fysisk_aktivitet.pdf", 
             "./data/Rapport-Raad-om-baeredygtig-kost.pdf", 
             "./data/Sund kost - Patienthåndbogen på sundhed.dk.pdf"]

for pdf_file in pdf_files:
    with pdfplumber.open(pdf_file) as pdf:
        pdf_text = ""
        for page in pdf.pages:
            pdf_text += page.extract_text()
        pdf_texts.append(pdf_text)
```

```python
print('%.1000s' % pdf_texts[0])
```

```python
# loading data from excel spreadsheet
excel_file = "./data/Frida_5.1_November_2023_new.xlsx"

excel_data = pd.read_excel(excel_file)

excel_text = " ".join(" ".join(str(cell) for cell in row) for _, row in excel_data.iterrows())
```

```python
# Printing the first 1000 char
print('%.1000s' % excel_text)
```

```python
import re
```

```python
# Cleaning the pdf text
def clean_pdf_text(text):
    text = text.replace('\n', ' ') # Replaces all \n with a whitespace
    text = re.sub(r'\.{2,}', ' ', text) # Replaces 2 or more consecutive periods with a whitespace
    text = re.sub(r'\s+', ' ', text) # Removes multpiple whitespaces in a row
    return text
    
texts = [clean_pdf_text(s) for s in pdf_texts]

print('%.1000s' % texts[0])
```

```python
# Cleaning the excel text
def clean_excel_text(text):
    text = text.replace('nan','') # Removes nan
    text = re.sub(r'\s+', ' ', text) # Removes multpiple whitespaces in a row
    return text

cleaned_excel_text = clean_excel_text(excel_text)

print('%.1000s' % cleaned_excel_text)
```

```python
# Add the pdf and excel text together

texts.append(cleaned_excel_text)

print('%.1000s' % texts[3])
```

```python
from langchain.llms import Ollama
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate
from langchain.callbacks.manager import CallbackManager
from langchain.callbacks.streaming_stdout import StreamingStdOutCallbackHandler
```

```python
llm = Ollama(model="llama3", callback_manager = CallbackManager([StreamingStdOutCallbackHandler()]))
```

```python
template = """Brug de forskellige tekst dokumenter til at svare på spørgsmålene. 
Hvis du ikke kan svare, bare sig du ikke ved det lad hver med at prøve at finde på et svar. 
Brug maks fem sætninger. Gør svaret så kort som muligt. 

{context}

Question: {question}

Helpful Answer:
"""
```

```python
prompt = PromptTemplate.from_template(template)
chain = RetrievalQA.from_chain_type(
    llm,
    retriever=vectordb.as_retriever(),
    return_source_documents=True,
    chain_type_kwargs={"prompt": prompt})
```

```python
question3 = 'Hvordan spiser man sundt?'
```

```python
result = chain({"query": question3})
result["result"]
```

```python
question4 = 'Hvor meget protein skal jeg have om dagen?'
```

```python
result = chain({"query": question4})
result["result"]
```

```python
question5 = 'Hvor mange grøntsager skal jeg spise hver dag?'
```

```python
result = chain({"query": question5})
result["result"]
```
